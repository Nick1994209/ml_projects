```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

# Гамма распределение
## https://en.wikipedia.org/wiki/Gamma_distribution
### теоретические формулы:
### мат ожидание = k * θ 
### дисперсия = k * (θ**2)

```python
# определяем распределение. Вводим константы мат ожидание M и дисперсию D
k = 2
theta = 3
distribution = sts.gamma(a=k,scale=theta)
M = 6.0  # теоретическое мат ожидание
D = 18.0 # теоретическая дисперсия
```

```python
sample = distribution.rvs(1000)
```

```python
# строю гистограмму выборки и рисую поверх неё теоретическую плотность распределения моей 
# случайной величины

a = round(min(sample))
b = round(max(sample))
N = len(sample)
x = np.linspace(a,b,N)
pdf = distribution.pdf(x)
plt.plot(x, pdf, label='theoretical probability density function')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.hist(sample,bins=37,normed=True,label='histogram of sample')
plt.legend()
```

```python
# эта функция генерит массив выборочного среднего случайной величины при разных объёмах выборок 
def generateSamplesOfN(distribution,n):
    retVal = list()
    for i in range(0,N):
        retVal.append(np.mean(distribution.rvs(n)))
    return retVal    
```

```python
# запускаю функцию для n = 5,10,50
sample5  = generateSamplesOfN(distribution,5)
sample10 = generateSamplesOfN(distribution,10)
sample50 = generateSamplesOfN(distribution,50)
```

```python
# Поверх гистограммы n = 5 рисую плотность  нормального распределения c мат ожиданием = исходному 
# и дисперсией равной (исходная дисперсия)/n
norm5= sts.norm(loc = M, scale = math.sqrt(D/5))
plt.plot(x,norm5.pdf(x),label="norm")
plt.hist(sample5,normed=True,label="hist")
plt.legend()
```

```python
norm10= sts.norm(loc = M, scale = math.sqrt(D/10))
plt.plot(x,norm10.pdf(x),label="norm")
plt.hist(sample10,normed=True,label="hist")
plt.legend()
```

```python
norm50= sts.norm(loc = M, scale = math.sqrt(D/50.0))
plt.plot(x,norm50.pdf(x),label="norm")
plt.hist(sample50,normed=True,label="hist")
plt.legend()
```

# Выводы
я вижу, что с ростом n точность аппроксимации растет. т.е визуально видно, что моя гистограмма все меньше отклоняется от нормального распределения с параметрами выбранными согластно ЦПТ. Также видно, что дисперсия стремится к нулю.


```python
plt.hist(sample5,bins=40,  label='histogram n = 5')
plt.hist(sample10,bins=40, label='histogram n = 10')
plt.hist(sample50,bins=40, label='histogram n = 50')
plt.legend()
```

## по гистограммам видно, что мат ожидание одинаково ,а дисперсия уменьшается
